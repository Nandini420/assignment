```python
import pandas as pd
import numpy as np
```

```python
flip = pd.read_csv("flip.csv", encoding= 'unicode_escape')
amaz = pd.read_csv("amz.csv", encoding= 'unicode_escape')
```

```python
flip.columns
```

```python
# droping the unwanted columns in flipkart dataset
flip = flip.drop(['uniq_id', 'crawl_timestamp', 'product_url',
       'product_category_tree', 'pid','image', 'is_FK_Advantage_product', 'description', 'product_rating',
       'overall_rating', 'brand', 'product_specifications'], axis=1)
```

```python
amaz.columns
```

```python
# droping the unwanted column in amazon dataset
amaz= amaz.drop(['uniq_id', 'crawl_timestamp', 'product_url',
       'product_category_tree', 'pid','image', 'is_FK_Advantage_product', 'description', 'product_rating',
       'overall_rating', 'brand', 'product_specifications'], axis=1)
```

```python
# based on product name merging the similar columns from  both the datasets
results = amaz.merge(flip, on="product_name", how="inner")
results.to_csv("results.csv")
```

```python
results = pd.read_csv("results.csv")
```

```python
results.head(20)
```

```python
#create duplicate product column for amazon
results['product name in amazon'] = results.loc[:, 'product_name']
```

```python
# deleting the unwanted column
results = results.drop(['Unnamed: 0'], axis=1)
```

```python
# changing the column names as required
results = results.rename(columns={'product_name':'product name in flipkart', 'retail_price_x':'retail price in flipkart', 
                        'retail_price_y':'retail price in amazon', 'discounted_price_x':'discounted price in flipkart',
                        'discounted_price_y':'discounted price amazon'})
```

```python
# to get the columns in required position
results = results[['product name in flipkart', 'retail price in flipkart', 'discounted price in flipkart', 
                   'product name in amazon','retail price in amazon','discounted price amazon']]
```
